```python
import pandas as pd

df = pd.read_csv("daraz_data.csv", encoding="utf-8-sig")

# Extract number and prefix (like 'k' or 'm') from itemSoldCntShow
df[["num", "prefix"]] = df["itemSoldCntShow"].str.extract(r'([\d.]+)([a-zA-Z]?)')

# Convert number part to float
df["num"] = pd.to_numeric(df["num"], errors="coerce")

# Normalize prefix to lowercase
df["prefix"] = df["prefix"].str.lower()

# Convert based on prefix
df["itemSoldCntShow"] = df["num"].where(df["prefix"] == "", df["num"])  # default copy
df.loc[df["prefix"] == "k", "itemSoldCntShow"] = df["num"] * 1000
df.loc[df["prefix"] == "m", "itemSoldCntShow"] = df["num"] * 1_000_000

# Drop helper columns
df.drop(columns=["num", "prefix"], inplace=True)

df.head()
```
